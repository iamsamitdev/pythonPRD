```python
----------------------------
ส่วนของคำสั่ง import
----------------------------
----------------------------------
# ฟังก์ชันเชื่อมต่อฐานข้อมูล
----------------------------------
--------------------------------------------------------------------
# ฟังก์ชัน subject สำหรับจัดการข้อมูลวิชาเรียน
---------------------------------------------------------------------
def subject():
    mainFrm.withdraw()  # เมื่อหน้าจอ subject ขึ้นมา หน้าจอหลักจะถูกปิด

    # ฟังก์ชันปิดหน้าต่างเฟรม subject
    def showMainFrm():
        frmSubject.destroy()   # ปิดหน้าจอใดๆ ที่เปิดอยู่
        mainFrm.update()       
        mainFrm.deiconify()    # แสดงหน้าจอหลัก

    # ฟังก์ชันลบข้อมูลออกจาก Entry widget
    def clearEntryOfSubject():
        entSubID.delete(0, END)
        entSubname.delete(0, END)
        entUnit.delete(0, END)
        entTeaID.delete(0, END)

    # ฟังก์ชันแสดงข้อมูลที่ Entry widget เมื่อคลิกที่แถว Treeview widget 
    def SelectRowTreeViewSubject(event):
        clearEntryOfSubject() # เรียกใช้ฟังก์ชันเพื่อล้างข้อมูลออกจาก Entry widget
        # กำหนดให้แสดงข้อมูลที่ Entry widget เมื่อคลิกที่แถวของ Treeview widget
        item = viewSub.selection()
        for i in item:
            entSubID.insert("", viewSub.item(i, "values")[0])
            entSubname.insert("", viewSub.item(i, "values")[1])
            entUnit.insert("", viewSub.item(i, "values")[2])
            entTeaID.insert("", viewSub.item(i, "values")[3])

    # ฟังก์ชันให้แสดงผลข้อมูลจากตาราง subject
    def insertDataToTreeView():
        # ลบข้อมูลออกจาก Treeview widgit
        for c in viewSub.get_children():
            viewSub.delete(c)
        con_db()  # เรียกใช้ฟังก์ชันเชื่อมต่อฐานข้อมูล
        cs.execute("SELECT * FROM subjects")
        data = cs.fetchall()
        # อ่านข้อมูลและเพิ่มข้อมูลไปยัง Treeview widget ทีละแถว
        for d in data:
            viewSub.insert("", "end", values=d)
        con.close()
        cs.close()

    # ฟังก์ชันเพิ่มข้อมูลเข้าตาราง subjects
    def insert_data():
        # ตรวจสอบ Entry widget มีค่าว่างหรือไม่
        if entSubID.get()=="" or entSubname.get() =="" or entUnit.get() == "" 
                              or entTeaID.get()=="":
            messagebox.showwarning("Information", "กรุณากรอกข้อมูลให้ครบด้วยครับ !!")
        else:
            con_db()    # เรียกใช้ฟังก์ชันเชื่อมต่อฐานข้อมูล
            #  คำสั่งเพิ่มข้อมูลลงตาราง subjects
            cs.execute("INSERT INTO subjects(subject_id, subject_name, unit, teacher_id)"\
                       "VALUES(%s, %s, %s, %s)"\
                       ,(str(entSubID.get()), str(entSubname.get()), 
                         str(entUnit.get()), str(entTeaID.get())))
            con.commit()      # คำสั่งปรับปรุงข้อมูลในฐานข้อมูล
            con.close()
            cs.close()
            clearEntryOfTeacher()       # เรียกใช้ฟังก์ชันเพื่อล้างข้อมูลออกจาก Entry widget
            insertDataToTreeView()      # เรียกใช้ฟังก์ชันเพื่อปรับปรุงข้อมูลใน Treeview widget
            # แสดง messageBox เตือนเมื่อบันทึกข้อมูลเสร็จ
            messagebox.showwarning("Information", "บันทึกข้อมูลเรียบร้อยแล้ว ครับ")  


    # ฟังก์ชันลบข้อมูลออกจากตาราง subjects
    def delete_data():
        if not viewSub.selection():
            messagebox.showinfo("Information", "คุณยังไม่เลือกข้อมูลที่ต้องการลบ")
        else:
            item = viewSub.selection()
            for i in item:
                con_db()      # เรียกใช้งานเชื่อมต่อฐานข้อมูล
               # คำสั่งลบข้อมูลออกจากตาราง teachers
                cs.execute("DELETE FROM subjects WHERE subject_id = %s " % 
                           viewSub.item(i, "values")[0])
                # ถ้าตอบ Yes เป็นการยืนยันการลบข้อมูล
                msgBox = messagebox.askquestion("Warning", "ต้องการลบข้อมูลนี้ใช่หรือไม่", 
                                                icon="warning")
                if msgBox == "yes":
                    con.commit()  # ยืนยันการลบข้อมูล
            con.close()
            cs.close()
            clearEntryOfTeacher()  # ลบข้อมูลใน Entry widget
            insertDataToTreeView() # ปรับปรุงข้อมูลใน Treeview widget

    # ฟังก์ชันปรับปรุงข้อมูลตาราง subjects
    def update_data():
        # กำหนดให้ Entry widget ของ entID แสดงผลข้อมูลได้อย่างเดียว
        entStyle.configure("entSubID.TEntry", state="readonly")     
        # แจ้งเตือนถ้ายังไม่มีการเลือกข้อมูลที่ต้องการแก้ไข
        if not viewTcr.selection():
            messagebox.showinfo("Warning", "คุณยังไม่เลือกข้อมูลที่ต้องการแก้ไข")
        else:
            con_db()
            cs.execute("UPDATE subjects set subject_id=%s, subject_name=%s,"\ 
                       "unit=%s, teacher_id=%s WHERE subject_id=%s"\
                       ,(str(entSubID.get()), str(entSubname.get()), 
                         str(entUnit.get()), str(entTeaID.get()), str(entSubID.get())))

        # แจ้งเตือนก่อนปรับปรุงข้อมูลด้วย messagebox ถ้าตอบ yes ข้อมูลจะถูกปรับปรุง
            msgBox = messagebox.askquestion("Information", "ต้องการแก้ไขข้อมูลนี้ใช่หรือไม่", 
                                            icon="warning")
            if msgBox == "yes":
                con.commit()
            con.close()
            cs.close()
            clearEntryOfTeacher()  # เรียกใช้ฟังก์ชันลบข้อมูลออกจาก Entry widget
            insertDataToTreeView() # เรียกใช้ฟังก์ชันเพื่อปรับปรุงข้อมูล Treeview widget

    # คำสั่งสร้าง Treeview widget
    frmSubject = Toplevel()     # กำหนดให้ปรากฏขึ้นมาเป็นหน้าต่างบนสุด
    frmSubject.title("อาจารย์ผู้สอน") # แสดงข้อความที่ title bar
    ttk.Frame(frmSubject, height=410, width=530).pack() # กำหนดขนาด Treeview widget
    frmSubject.resizable(width=0, height=0) # กำหนดไม่ให้สามารถปรับขนาดได้

    # กำหนดขนาดและจำนวนคอลัมน์ Treeview widget
    viewSub = ttk.Treeview(frmSubject, column=("Sub_ID", "Sub_name", "unit", "teact_ID"), 
                           show="headings", selectmode="extended")
    viewSub.place(height=200, width=500, x=10, y=200)

    # กำหนดให้แสดง scrollbar 
    sc = ttk.Scrollbar(frmSubject, orient="vertical", command=viewSub.yview)
    sc.place(height=200, x=500, y=200)    # กำหนดความยาวและตำแหน่ง scrollbar
    viewSub.configure(yscrollcommand=sc.set)

    # กำหนดขนาดความกว้างของคอลัมน์ Treeview widget
    viewSub.column("#1", width="30")
    viewSub.column("#2", width="150")
    viewSub.column("#3", width="50")
    viewSub.column("#4", width="50") 

    # กำหนดที่หัวแต่ละคอลัมน์ 
    viewSub.heading("Sub_ID", text="รหัสวิชา")
    viewSub.heading("Sub_name", text="ชื่อวิชา")
    viewSub.heading("unit", text="จำนวนหน่วยกิต")
    viewSub.heading("teact_ID", text="รหัสอาจารย์ผู้สอน")
    # กำหนดให้เมื่อคลิกที่ Treeview widget ข้อมูลแต่ละคอลัมน์จะแสดงผลที่ 
    # Entry widget ผ่านฟังก์ชัน SelectTreeTeacher
    viewTcr.bind('<ButtonRelease>', SelectRowTreeViewSubject)  
    insertDataToTreeView() # เรียกใช้ฟังก์ชันแสดงข้อมูลใน TreeView widget ทันทีเมื่อแสดงผลหน้าต่างอาจารย์

    # สร้างเฟรม label widget
    ttk.LabelFrame(frmSubject, text="เพิ่มหรือแก้ไขข้อมูลวิชาเรียน").place(height=185, width=400, 
                                                                   x=10, y=10) 

    # สร้าง label widget และกำหนดตำแหน่ง
    ttk.Label(frmSubject, text="รหัสวิชา :").place(height=20, width=60, x=20, y=40)
    ttk.Label(frmSubject, text="ชื่อวิชา :").place(height=20, width=60, x=20, y=70)
    ttk.Label(frmSubject, text="จำนวนหน่วยกิต :").place(height=20, width=80, x=20, y=100)
    ttk.Label(frmSubject, text="อาจารย์ผู้สอน :").place(height=20, width=80, x=20, y=130) 

    # สร้าง Entry widget 
    entStyle = ttk.Style()
    entSubID = ttk.Entry(frmSubject, style="entID.TEntry")
    entSubname = ttk.Entry(frmSubject)
    entUnit = ttk.Entry(frmSubject)
    entTeaID = ttk.Entry(frmSubject) 

    # กำหนดตำแหน่งและขนาด entry widget
    entSubID.place(height=25, width=100, x=110, y=40)
    entSubname.place(height=25, width=200, x=110, y=70)
    entUnit.place(height=25, width=200, x=110, y=100)
    entTeaID.place(height=25, width=100, x=110, y=130)

    # กำหนดรูปแบบแสดงผลบน button widget
    btnStyle = ttk.Style()
    btnStyle.configure("btnSave.TButton", font=("Times", "10", "bold"))
    btnStyle.configure("btnEdit.TButton", font=("Times", "10", "bold"))
    btnStyle.configure("btnDelete.TButton", font=("Times", "10", "bold"))
    btnStyle.configure("btnClear.TButton", font=("Times", "10", "bold"))
    btnStyle.configure("btnExit.TButton", font=("Times", "10", "bold"))

    # สร้าง button widget
    btnSave = ttk.Button(frmSubject, text="บันทึก", command=insert_data, 
                         style="btnSave.TButton")
    btnEdit = ttk.Button(frmSubject, text="แก้ไข", command=udate_data, 
                         style="btnEdit.TButton")
    btnDelete = ttk.Button(frmSubject, text="ลบ", command=delete_data, 
                           style="btnDelete.TButton")
    btnCancle = ttk.Button(frmSubject, text="ยกเลิก", command=clearEntryOfSubject, 
                           style="btnClear.TButton")
    btnClose = ttk.Button(frmSubject, text="ปิดโปรแกรม", command=showMainFrm, 
                          style="btnExit.TButton")

    # กำหนดตำแหน่งและขนาด button widget
    btnSave.place(height=35, width=100, x=420, y=20)
    btnEdit.place(height=35, width=100, x=420, y=55)
    btnDelete.place(height=35, width=100, x=420, y=90)
    btnCancle.place(height=35, width=100, x=420, y=125)
    btnClose.place(height=35, width=100, x=420, y=160)
    
------------------------------------------
ส่วนของคำสั่งแสดงผลหน้าเมนูหลัก
------------------------------------------
mainFrm.mainloop()
```
